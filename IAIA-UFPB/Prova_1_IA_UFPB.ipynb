```python
# columns were extracted from http://kdd.ics.uci.edu/databases/kddcup99/task.html
columns = columns = ["duration", "protocoltype", "service", "flag", "srcbytes", "dstbytes", "land", "wrongfragment",
           "urgent", "hot", "numfailedlogins", "loggedin", "numcompromised", "rootshell", "suattempted",
           "numroot", "numfilecreations", "numshells", "numaccessfiles", "numoutboundcmds", "ishostlogin",
           "isguestlogin", "count", "srvcount", "serrorrate", "srvserrorrate", "rerrorrate", "srvrerrorrate",
           "samesrvrate", "diffsrvrate", "srvdiffhostrate", "dsthostcount", "dsthostsrvcount",
           "dsthostsamesrvrate", "dsthostdiffsrvrate", "dsthostsamesrcportrate", "dsthostsrvdiffhostrate",
           "dsthostserrorrate", "dsthostsrvserrorrate", "dsthostrerrorrate", "dsthostsrvrerror_rate", "labels"]

# data acquisition           
df = pd.read_csv("kddcup.data_10_percent", names=columns)
df.info()
df.describe()
```

```python
# categorical data
cat = pd.get_dummies(df.iloc[:,1:4])

# attributes and target
X = pd.concat([df.iloc[:,0], df.iloc[:, 4:41], cat], axis=1)
y = df.iloc[:, 41]

# dimensionality reduction (PCA)
pca = PCA(n_components=20)
pca.fit(X)
X = pca.transform(X)

# data splitting
X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y)
```

```python
# model building

knn = KNeighborsClassifier(metric="manhattan", n_neighbors=5)
knn.fit(X_train, y_train)
print("\nAcuracy: ", knn.score(X_test, y_test)*100, "%")
print("\nAvrg. Precision: ", sk.metrics.average_precision_score(y_test, knn.predict(X_test)))
print("\nF1 Score: ", sk.metrics.f1_score(y_test, knn.predict(X_test)))
```

```python
# (4) KNN aplicado ao dataset da questão 3
# Dataset Extracted from: K. R. Purba, D. Asirvatham and R. K. Murugesan, "Classification of instagram fake users using supervised machine learning algorithms," International Journal of Electrical and Computer Engineering (IJECE), vol. 10, no. 3, pp. 2763-2772, 2020.

# data acquisition
df = pd.read_csv("https://raw.githubusercontent.com/J0AZZ/artificial-intelligence_studies/master/IAIA-UFPB/user_fake_authentic_2class.csv")
df.info()
df.describe()

# input and target attribution
y = df.iloc[:,17]
x = df.iloc[:,:17]

# data normalization
normalized_x = pd.DataFrame(sk.preprocessing.normalize(x))

# training and validation splitting
x_train, x_test = normalized_x.iloc[:50000, :], normalized_x.iloc[50000:, :]
y_train, y_test = y.iloc[:50000], y.iloc[50000:]

# model building with two different metrics and 5 values of k
for metric in ["manhattan", "euclidean"]:
  print("\nUsing metric ", metric)
  for k in range(1,6):
    knn = KNeighborsClassifier(n_neighbors=k, metric=metric, weights='distance')
    knn.fit(x_train, y_train)
    print("Accuracy: ", knn.score(x_test, y_test)*100, "%")
    print("\nAvrg. Precision: ", sk.metrics.average_precision_score(y_test, knn.predict(X_test)))
    print("\nF1 Score: ", sk.metrics.f1_score(y_test, knn.predict(X_test)))
```
